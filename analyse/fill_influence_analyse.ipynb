```python
import os

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from statannotations.Annotator import Annotator
```

```python
def violinplot(data, x=None, y=None, hue=None, pairs=None, xlabel=None, ylabel=None, xticklabels=None, output_path: str = None, figsize: tuple = (24, 12)):
    sns.set_theme(style="whitegrid")

    fig, ax = plt.subplots(figsize=figsize)

    ax = sns.violinplot(data=data, x=x, y=y, hue=hue)

    if pairs:
        annotator = Annotator(ax, pairs, data=data, x=x, y=y, hue=hue)
        annotator.configure(test='Mann-Whitney', text_format='star')
        annotator.apply_and_annotate()

    plt.legend()
    plt.grid(False)

    if xlabel is not None:
        plt.xlabel(xlabel)
    if ylabel is not None:
        plt.ylabel(ylabel)
    if xticklabels:
        ax.set_xticklabels(labels=xticklabels)

    if output_path:
        plt.savefig(output_path, bbox_inches="tight")
    plt.show()
```

```python
def read_accuracy(dir_path: str):
    acc = {}
    for target in ['minor', 'important', 'full', 'all']:
        acc[target] = pd.read_csv(os.path.join(dir_path, 'accuracy_{}.csv'.format(target)), index_col=0)
    return acc

control_acc = read_accuracy('../output/control')
miNNseq_acc = read_accuracy('../output/miNNseq')
```

```python
for target in ['minor', 'important', 'full', 'all']:
    _control_acc = pd.melt(control_acc[target], value_vars=control_acc[target].columns)
    _miNNseq_acc = pd.melt(miNNseq_acc[target], value_vars=miNNseq_acc[target].columns)

    _control_acc['category'] = 'control'
    _miNNseq_acc['category'] = 'miNNseq'

    pairs = [
        ((method_name, 'control'), (method_name, 'miNNseq'))
        for method_name in control_acc[target].columns
    ]
    pairs.extend([
        (('RandomForestClassifier', 'control'), ('ExtraTreesClassifier', 'control')),
        (('RandomForestClassifier', 'miNNseq'), ('ExtraTreesClassifier', 'miNNseq')),
    ])

    violinplot(
        pd.concat([_control_acc, _miNNseq_acc]),
        x='variable',
        y='value',
        hue='category',
        pairs=pairs,
        xlabel='',
        ylabel='Accuracy',
        output_path='../output/miNNseq/accuracy_comparison({}).pdf'.format(target),
    )
```

```python
def read_roc(dir_path: str):
    roc = {}
    for target in ['minor', 'important', 'full', 'all']:
        roc[target] = np.load(os.path.join(dir_path, 'ROC_{}.npy'.format(target)), allow_pickle=True)
    return roc

control_roc = read_roc('../output/control')
miNNseq_roc = read_roc('../output/miNNseq')
```

```python
for target in ['minor', 'important', 'full', 'all']:
    _control_roc = control_roc[target].item()
    _miNNseq_roc = miNNseq_roc[target].item()

    _control_roc = pd.DataFrame(
        [
            _control_roc[method_name][2]
            for method_name in _control_roc.keys()
        ],
        index=_control_roc.keys()
    ).T
    _miNNseq_roc = pd.DataFrame(
        [
            _miNNseq_roc[method_name][2]
            for method_name in _miNNseq_roc.keys()
        ],
        index=_miNNseq_roc.keys()
    ).T

    _control_roc = pd.melt(_control_roc, value_vars=_control_roc.columns)
    _miNNseq_roc = pd.melt(_miNNseq_roc, value_vars=_miNNseq_roc.columns)

    _control_roc['category'] = 'control'
    _miNNseq_roc['category'] = 'miNNseq'

    pairs = [
        ((method_name, 'control'), (method_name, 'miNNseq'))
        for method_name in control_acc[target].columns
    ]
    pairs.extend([
        (('RandomForestClassifier', 'control'), ('ExtraTreesClassifier', 'control')),
        (('RandomForestClassifier', 'miNNseq'), ('ExtraTreesClassifier', 'miNNseq')),
    ])

    violinplot(
        pd.concat([_control_roc, _miNNseq_roc]),
        x='variable',
        y='value',
        hue='category',
        pairs=pairs,
        xlabel='',
        ylabel='ROC-AUC',
        output_path='../output/miNNseq/auc_comparison({}).pdf'.format(target),
    )
```
