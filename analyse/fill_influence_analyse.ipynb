```python
import os

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from statannotations.Annotator import Annotator
```

```python
def violinplot(data, x=None, y=None, hue=None, pairs=None, xticklabels=None, output_path: str = None, figsize: tuple = (24, 12)):
    sns.set_theme(style="whitegrid")

    fig, ax = plt.subplots(figsize=figsize)

    ax = sns.violinplot(data=data, x=x, y=y, hue=hue)

    if pairs:
        annotator = Annotator(ax, pairs, data=data, x=x, y=y, hue=hue)
        annotator.configure(test='Mann-Whitney', text_format='star')
        annotator.apply_and_annotate()

    plt.legend()
    plt.grid(False)

    if x:
        plt.xlabel(x)
        plt.xticks(rotation=90)
    if y:
        plt.ylabel(y)
    if xticklabels:
        ax.set_xticklabels(labels=xticklabels)

    if output_path:
        plt.savefig(output_path, bbox_inches="tight")
    plt.show()
```

```python
def read_acc(dir_path: str):
    return {
        'effective': pd.read_csv(os.path.join(dir_path, 'accuracy_effective.csv'), header=[0], index_col=0),
        'markedly_effective': pd.read_csv(os.path.join(dir_path, 'accuracy_markedly effective.csv'), header=[0], index_col=0),
        'cured': pd.read_csv(os.path.join(dir_path, 'accuracy_cured.csv'), header=[0], index_col=0),
        'all': pd.read_csv(os.path.join(dir_path, 'accuracy_all.csv'), header=[0], index_col=0),
    }

control_acc = read_acc('../output/control')
KNN_acc = read_acc('../output/data_fill/KNN')
rpca_acc = read_acc('../output/data_fill/rpca')
```

```python
for target in ['effective', 'markedly_effective', 'cured', 'all']:
    _control_acc = pd.melt(control_acc[target], value_vars=control_acc[target].columns)
    _KNN_acc = pd.melt(KNN_acc[target], value_vars=KNN_acc[target].columns)

    _control_acc['category'] = 'control'
    _KNN_acc['category'] = 'KNN'

    violinplot(
        pd.concat([_control_acc, _KNN_acc]),
        x='variable',
        y='value',
        hue='category',
        pairs=[
            (('AdaBoostClassifier', 'control'), ('AdaBoostClassifier', 'KNN')),
            (('GradientBoostingClassifier', 'control'), ('GradientBoostingClassifier', 'KNN')),
            (('RandomForestClassifier', 'control'), ('RandomForestClassifier', 'KNN')),
            (('ExtraTreesClassifier', 'control'), ('ExtraTreesClassifier', 'KNN')),
            (('SupervisedDBNClassification', 'control'), ('SupervisedDBNClassification', 'KNN')),
            (('SVC', 'control'), ('SVC', 'KNN')),
        ],
        output_path='../output/data_fill/KNN/effect_comparison({}).pdf'.format(target),
    )
```

```python
def boxplot(data, x=None, y=None, hue=None, pairs=None, xticklabels=None, output_path: str = None, figsize: tuple = (24, 12)):
    sns.set_theme(style="whitegrid")

    fig, ax = plt.subplots(figsize=figsize)

    ax = sns.boxplot(data=data, x=x, y=y, hue=hue)

    if pairs:
        annotator = Annotator(ax, pairs, data=data, x=x, y=y, hue=hue)
        annotator.configure(test='Mann-Whitney', text_format='star')
        annotator.apply_and_annotate()

    plt.legend()
    plt.grid(False)

    if x:
        plt.xlabel(x)
        plt.xticks(rotation=90)
    if y:
        plt.ylabel(y)
    if xticklabels:
        ax.set_xticklabels(labels=xticklabels)

    if output_path:
        plt.savefig(output_path, bbox_inches="tight")
    plt.show()
```
